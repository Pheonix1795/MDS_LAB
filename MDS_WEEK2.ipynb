## 1. Indexing NumPy Arrays

### 1.1 Accessing Single Elements

```python
# Assumes 'arr' is defined in the preceding cell.
print(f"Element at index 0: {arr[0]}")
```

### 1.2 Accessing Elements in 2D Arrays

```python
# Assumes 'arr_2d' is defined in the preceding cell.
print(f"Element at row 0, column 1: {arr_2d[0, 1]}")
print(f"Element at row 2, column 0: {arr_2d[2, 0]}")
```

### 1.3 Boolean Indexing

```python
# Assumes 'arr' is defined in the preceding cell.
print(f"Elements greater than 5 in 1D array: {arr[arr > 5]}")
```

## 2. Slicing NumPy Arrays

### 2.1 Basic 1D Slicing

```python
# Assumes 'arr' is defined in the preceding cell.
print(f"Slice from index 2 to 5: {arr[2:5]}")
print(f"Slice from beginning to index 4: {arr[:4]}")
print(f"Slice from index 6 to end: {arr[6:]}")
print(f"Slice with step 2: {arr[::2]}")
```

### 2.2 2D Array Slicing

```python
# Assumes 'arr_2d' is defined in the preceding cell.
print(f"Slice rows 0-1, columns 1-2:\n{arr_2d[0:2, 1:3]}")
```

## 3. Reshaping NumPy Arrays

### 3.1 Reshaping to Specific Dimensions

```python
# Assumes 'arr' is defined in the preceding cell.
arr_reshaped = arr.reshape(3, 4)
print(f"Reshaped to 3x4:\n{arr_reshaped}")
arr_reshaped_2 = arr.reshape(2, 6)
print(f"Reshaped to 2x6:\n{arr_reshaped_2}")
```

### 3.2 Flattening Arrays

```python
# Assumes 'arr_reshaped' is defined in the preceding sub-section or main cell.
arr_flat = arr_reshaped.flatten()
print(f"Flattened array: {arr_flat}")
```

## 4. Joining NumPy Arrays

### 4.1 Concatenation

```python
# Assumes arr1, arr2, arr3, arr4 are defined in the preceding cell.
arr_joined_1d = np.concatenate((arr1, arr2))
print(f"Concatenated 1D (np.concatenate): {arr_joined_1d}")
arr_joined_rows = np.concatenate((arr3, arr4), axis=0)
print(f"Concatenated along axis 0 (rows):\n{arr_joined_rows}")
arr_joined_cols = np.concatenate((arr3, arr4), axis=1)
print(f"Concatenated along axis 1 (columns):\n{arr_joined_cols}")
```

### 4.2 Stacking

```python
# Assumes arr3, arr4 are defined in the preceding cell.
arr_vstacked = np.vstack((arr3, arr4))
print(f"Vertically stacked (np.vstack):\n{arr_vstacked}")
arr_hstacked = np.hstack((arr3, arr4))
print(f"Horizontally stacked (np.hstack):\n{arr_hstacked}")
```

## 5. Splitting NumPy Arrays

### 5.1 Horizontal Splitting (np.hsplit)

```python
# Assumes 'arr_to_split' is defined in the preceding cell.
split_h = np.hsplit(arr_to_split, 2)
print(f"Horizontal split (2 parts):")
for i, part in enumerate(split_h):
    print(f"Part {i+1}:\n{part}")
```

### 5.2 Vertical Splitting (np.vsplit)

```python
# Assumes 'arr_to_split' is defined in the preceding cell.
split_v = np.vsplit(arr_to_split, 4)
print(f"Vertical split (4 parts):")
for i, part in enumerate(split_v):
    print(f"Part {i+1}:\n{part}")
```
