````python
import numpy as np
import time

# Data initialization
l1 = [2, 4, 6, 8, 10]
l2 = [1, 2, 3, 4, 5]

# 1. Measuring Python List Operations (using list comprehension)
start_list = time.time()
sum_list = [x + y for x, y in zip(l1, l2)]
sub_list = [x - y for x, y in zip(l1, l2)]
mul_list = [x * y for x, y in zip(l1, l2)]
div_list = [x / y for x, y in zip(l1, l2)]
end_list = time.time()
duration_list = end_list - start_list

# 2. Measuring NumPy Operations
arr_l1 = np.array(l1)
arr_l2 = np.array(l2)

start_np = time.time()
sum_np = arr_l1 + arr_l2
sub_np = arr_l1 - arr_l2
mul_np = arr_l1 * arr_l2
div_np = arr_l1 / arr_l2
end_np = time.time()
duration_np = end_np - start_np

print(f"Python List operations took: {duration_list:.10f} seconds")
print(f"NumPy Array operations took: {duration_np:.10f} seconds")
print("\nResult (NumPy Sum):", sum_np)
````
