## Section 2: Performance Benchmarking
### 2.1 Time Complexity: Python Lists vs. NumPy

```python
import time
import numpy as np

# Size of the data
size = 1000000

# Python List Performance
list_data = list(range(size))
start_time = time.time()
list_result = [x**2 for x in list_data]
python_time = time.time() - start_time
print(f"Python List time: {python_time:.5f} seconds")

# NumPy Array Performance
numpy_data = np.arange(size)
start_time = time.time()
numpy_result = numpy_data**2
numpy_time = time.time() - start_time
print(f"NumPy Array time: {numpy_time:.5f} seconds")

print(f"\nNumPy is approximately {python_time / numpy_time:.2f}x faster!")
```

### 2.2 Memory Efficiency: Python Lists vs. NumPy

```python
import sys
import numpy as np

# Define the number of elements
n = 1000000

# Memory usage for a Python list
# A list stores pointers to objects, so we account for the list overhead plus the size of each integer object
python_list = list(range(n))
list_memory = sys.getsizeof(python_list) + sum(sys.getsizeof(i) for i in python_list)

# Memory usage for a NumPy array
numpy_array = np.arange(n, dtype=np.int64)
# nbytes returns the total bytes consumed by the elements of the array
numpy_memory = numpy_array.nbytes

print(f"Memory used by Python list: {list_memory / (1024**2):.2f} MB")
print(f"Memory used by NumPy array: {numpy_memory / (1024**2):.2f} MB")
print(f"\nNumPy uses approximately {list_memory / numpy_memory:.2f}x less memory!")
```

### 2.3 Operation Speed: Summation Benchmark

```python
import timeit
import numpy as np

# Configuration
size = 1000000
iterations = 10

# Python List calculation setup
list_data = list(range(size))
def python_sum():
    return sum(list_data)

# NumPy Array calculation setup
numpy_data = np.arange(size)
def numpy_sum():
    return np.sum(numpy_data)

# Run benchmarks
python_time = timeit.timeit(python_sum, number=iterations) / iterations
numpy_time = timeit.timeit(numpy_sum, number=iterations) / iterations

print(f"Average time for Python sum: {python_time:.6f} seconds")
print(f"Average time for NumPy sum:  {numpy_time:.6f} seconds")
print(f"\nNumPy is {python_time / numpy_time:.2f}x faster for summation!")
```

### 8.2 Joining Arrays
Joining means putting contents of two or more arrays in a single array. In NumPy, we join arrays by axes.

```python
# ### 8.2.1 Concatenating Arrays
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

# Join two 1D arrays
joined_arr = np.concatenate((arr1, arr2))
print("Concatenated 1D:", joined_arr)

# Join 2D arrays along an axis
arr2d_1 = np.array([[1, 2], [3, 4]])
arr2d_2 = np.array([[5, 6], [7, 8]])

joined_axis0 = np.concatenate((arr2d_1, arr2d_2), axis=0)
print("\nConcatenated 2D (Axis 0):\n", joined_axis0)
```

```python
# ### 8.2.2 Stacking Arrays
# Stacking is same as concatenation, the only difference is that stacking is done along a new axis.
s_arr1 = np.array([1, 2, 3])
s_arr2 = np.array([4, 5, 6])

# Stack along rows (hstack)
h_stack = np.hstack((s_arr1, s_arr2))
# Stack along columns (vstack)
v_stack = np.vstack((s_arr1, s_arr2))

print("Horizontal Stack:", h_stack)
print("Vertical Stack:\n", v_stack)
```

### 8.3 Splitting Arrays
Splitting is the reverse operation of Joining. We use `array_split()` for splitting arrays.

```python
# ### 8.3.1 Splitting a 1D Array
split_arr = np.array([1, 2, 3, 4, 5, 6])

# Split the array into 3 parts
newarr = np.array_split(split_arr, 3)

print("Original Array:", split_arr)
print("Split Array:", newarr)
print("First Part:", newarr[0])
```

## Section 10: Image Manipulation
### 10.1 Loading, Cropping, and Flipping Images

```python
# ## Section 10: Image Manipulation with NumPy
# ### 10.1 Loading, Cropping, and Flipping Images

import numpy as np
import matplotlib.pyplot as plt
from skimage import data

# Load a sample image (comes as a NumPy array)
img = data.astronaut()

# 1. Crop the image: [rows, columns]
# Selecting a region of interest (the astronaut's face/helmet area)
crop_img = img[30:230, 150:350]

# 2. Flip the image: [start:stop:step]
# Reversing the order of rows flips it vertically
flip_img = img[::-1, :, :]

# Displaying the results
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(img)
axes[0].set_title("Original")
axes[1].imshow(crop_img)
axes[1].set_title("Cropped (NumPy Slicing)")
axes[2].imshow(flip_img)
axes[2].set_title("Flipped (Step Indexing)")

for ax in axes: ax.axis("off")
plt.show()
```
